```python
#list length
mylist =["cat", "dog", "mouse"]
print(len(mylist))
```

```python
#A list with strings, integers and boolean values
mylist = ["cat", "dog", "mouse", 1, 2, 3, True, False]
print(mylist)
```

```python
#Access items
mylist = ["cat", "dog", "mouse"]
print(mylist[2])
```

```python
#reverse/negative indexing
mylist = ["cat", "dog", "mouse"]
print(mylist[-1])
```

```python
#slicing
mylist = ["cat", "dog", "mouse","elephant", "lion","peacock"]
print(mylist[2:5])
print(mylist[-1:-4:-1])
```

```python
#Append items
mylist = ["cat", "dog", "mouse"]
mylist.append("human")
mylist.append("67")
print(mylist)
```

```python
#Extend list
mylist = ["cat", "dog", "mouse"]
mylist.extend(["human", "67"])
print(mylist)
```

```python
#Insert items
mylist = ["cat", "dog", "mouse"]
mylist.insert(1,"human")
print(mylist)
```

```python
#Student Marks Management System
Student_Marks = [88,78,97,43,45,8,19,67,12,3]
print(Student_Marks)
highest = max(Student_Marks)
print(highest)
lowest = min(Student_Marks)
print(lowest)
Student_Marks.sort()
print(Student_Marks)
Student_Marks.sort(reverse=True)
print(Student_Marks)
```
