```python
# 1.Tokenization
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
#Word tokenization

from nltk.tokenize import word_tokenize

text = "This is a sample sentence for tokenization."
tokens = word_tokenize(text)
print(tokens)
```

```python
#Sentence tokanization

from nltk.tokenize import sent_tokenize

text = "This is a sample sentence. It has more than one sentence"
sentences = sent_tokenize(text)
print(sentences)
```

```python
#Stop word removal - words like is, the, and , in are removed

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
words = word_tokenize(text)
filtered_words = [word for word in words if word.lower() not in stop_words]

print(filtered_words)
```

```python
#Stemming - reduces words to their root form

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
words = ["running", "runs", "ran", "played", "playing"]

for w in words:
  print(stemmer.stem(w))
```

```python
#Lemmatization - coverts words to their dictionary form

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("swims"))
print(lemmatizer.lemmatize("runs"))
print(lemmatizer.lemmatize("plays"))
```

```python
#Part of speech (POS) tagging - identifies grammatical roles of words

import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag

words = word_tokenize("This is a sample sentence for POS tagging.")
pos_tags = pos_tag(words)

print(pos_tags)
```
