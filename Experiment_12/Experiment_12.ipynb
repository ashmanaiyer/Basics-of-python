```python
data={
    "Roll No":[101,102,103,104,105],
    "Name":["Asha","Rohan","Meera","Karan","Anil"],
    "Age":[20,"-",21,20,"22"],
    "Marks":[85,90,"-",88,76],
    "Department":["CSE","IT","CSE","cse","E&TC"],
    "Admission_Date":["01-06-2023","03-06-2023","05-06-2023","07-06-2023","10-06-2023"]

}

df=pd.DataFrame(data)
df
#Save as CSV
df.to_csv("Student_Data.csv",index=False)
```

```python
# Step 2: Replace '-' with NanN to mark missing values
df.replace("-",np.nan,inplace=True)
df
```

```python
#convert columns to numeric where needed
df["Age"]=pd.to_numeric(df["Age"],errors="coerce")
df["Marks"]=pd.to_numeric(df["Marks"],errors="coerce")
df
```

```python
#Step 3: Check Missing Values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
#Fill Missing age with mean
df["Age"].fillna(df["Age"].mean(),inplace=True)
df
#Fill missing marks with median
df["Marks"].fillna(df["Marks"].median(),inplace=True)
df
```

```python
#Step 5: Fix inconsistent text formatting
df["Department"]=df["Department"].str.upper()
df
```

```python
#Step 6: Convert admission date to standard datetime format
df["Admission_Date"]=pd.to_datetime(df["Admission_Date"])
df
```

```python
#Step 7: Save cleaned dataset
df.to_csv("claned_student_data.csv",index=False)
print("\nCleaned CSV file saved as cleaned_student_data.csv")
```

```python
#Working with missing valus o dataset-2
df=pd.read_csv('/content/Cars93 (1).csv')
df
```

```python
df.isna().sum()
```

```python
#imputing the missing value
#1 luggage room
df['Luggage.room']=df['Luggage.room'].fillna(df['Luggage.room'].mean())
df
```

```python
#2. Rear.seat.room
df['Rear.seat.room']=df['Rear.seat.room'].fillna(value=df['Rear.seat.room'].mean())
df
```

```python
#3.AirBags
df['AirBags']=df['AirBags'].fillna(value=df['AirBags'].mode().values[0])
df
```

```python
#Save cleaned datase
df.to_csv("Cleaned_Cars93.csv",index=False)
```
