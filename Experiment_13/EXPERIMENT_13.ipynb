```python
import pandas as pd
import numpy as np
data={
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price':[55000,20000,30000,2000,8000,45000],
    'Units_Sold':[25,60,35,80,50,15]
}

df=pd.DataFrame(data)
print(df)
```

```python
#Data Binning
bins1=[0,10000,30000,60000]
labels1=['Low','Medium','High']
df['Price_Category']=pd.cut(df['Price'],bins=bins1,labels=labels1)
print(df)
```

```python
#Binning Units Sold
bins1=[0,30,60,100]
labels1=['Less sale','Moderate sale','High sale']
df['Sales_Category']=pd.cut(df['Units_Sold'],bins=bins1,labels=labels1)
print(df)
```

```python
#Data Formatting
#Convert Data Type
print(df.dtypes)
df['Price']=df['Price'].astype(float)
print(df.dtypes)
```

```python
#Convert Text to Uppercase
df['Product']=df['Product'].str.upper()
print(df)
```

```python
#Rounding Values
df['Price']=df['Price'].round(2)
print(df)
```

```python
#Additional Operations
#Sort Data
sorted_df=df.sort_values(by='Price')
print(sorted_df)
```

```python
#Display Unique Values
print(df['Price_Category'].unique())
```

```python
#Dataset: Food Delivery Orders
import pandas as pd
import numpy as np
data={
    'Order_ID':['01','02','03','04','05','06','07','08'],
    'Order_Value':[250,450,120,600,300,150,500,220],
    'Delivery_Time':[30,45,25,60,35,20,50,28],
    'Distance_km':[5,8,3,10,6,2,9,4]
}
df=pd.DataFrame(data)
print(df)
```

```python
#Binning Order Value
bins=[0,200,400,700]
labels=['Low Order','Medium Order','High Order']
df['Order_Category']=pd.cut(df['Order_Value'],bins=bins,labels=labels)
print(df)
```

```python
#Binning Delivery Time
bins2=[0,30,45,70]
labels2=['Fast','Normal','Slow']
df['Delivery_Speed']=pd.cut(df['Delivery_Time'],bins=bins2,labels=labels2)
print(df)
```

```python
#Data Formatting Examples
#Convert Distance to Float
df['Distance_km']=df['Distance_km'].astype(float)
print(df)
```

```python
#Convert Order_Category to Uppercase
df['Order_Category']=df['Order_Category'].str.upper()
print(df)
```

```python
#Check Data Types
print(df.dtypes)
```

```python
#Unique Categories
print(df['Order_Category'].unique())
df
```

```python
#Count of categories
print(df['Order_Category'].value_counts())
df
```

```python
#Sorting
sorted_df=df.sort_values(by='Order_Value')
print(sorted_df)
df
```
