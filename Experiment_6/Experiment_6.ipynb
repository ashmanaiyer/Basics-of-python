```python
#5. Get input from user for 5 subjects and first average
#Calculate grade using if-elif-else
sub1=int(input("Enter a number:"))
sub2=int(input("Enter a number:"))
sub3=int(input("Enter a number:"))
sub4=int(input("Enter a number:"))
sub5=int(input("Enter a number:"))

avg=(sub1+sub2+sub3+sub4+sub5)/5
print("The average of the 5 sunjects is", avg)

if avg>=90:
  print("Grade O")
elif 90< avg >=80:
  print("Grade A")
elif 80< avg >=70:
  print("Grade B")
elif 70< avg >=60:
  print("Grade C")
elif 60 < avg >=50:
  print("Grade D")

else:
  print("Grade F")
```

```python
'''Python Program to check if a Date is Valid
and Print the incremented Date if valid'''
day = int(input("Enter a day: "))
month = int(input("Enter a month: "))
year = int(input("Enter a year: "))

if not (1 <= month <= 12):
        print("Date is invalid: Month must be between 1 and 12.")
else:
        if month in (1, 3, 5, 7, 8, 10, 12):
            max_days = 31
        elif month in (4, 6, 9, 11):
            max_days = 30
        elif month == 2:
            if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
                max_days = 29
            else:
                max_days = 28

        if not (1 <= day <= max_days):
            print("Date is invalid")
        else:
            print("Date is valid.")

            new_day = day + 1
            new_month = month
            new_year = year

            if new_day > max_days:
                new_day = 1
                new_month += 1
            if new_month > 12:
                new_month = 1
                new_year += 1

            print(f"Incremented Date: {new_day}/{new_month}/{new_year}")
```

```python
#Gross salary calculator
basic_sal = int(input("Enter the Basic Salary: "))

if(basic_sal <= 10000):
  HRA = (20/100) * basic_sal
  DA = (80/100) * basic_sal

elif(basic_sal <= 20000):
  HRA = (25/100) * basic_sal
  DA = (90/100) * basic_sal

else:
  HRA = (30/100) * basic_sal
  DA = (95/100) * basic_sal

gross_sal = basic_sal + HRA + DA
print(gross_sal)
```

```python
salary = int(input("Enter the salary: "))

if(salary <= 250000):
  tax = 0
elif(salary <= 500000):
  tax = (salary - 250000) * 0.05

elif(salary <= 1000000):
  tax = (250000 * 0.05) + (500000 * 0.20) + (salary - 1000000) * 0.30

print("The Tax amount to be paid is ", tax)
print("The salary In - Hand is ", salary - tax)
```
