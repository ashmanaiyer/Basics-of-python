```python

#Create DataFrame
import pandas as pd
data ={
    "Name":["A","B","C","D","E","F","G","H","I","J"],
    "Marks":[85,90,78,92,88,75,95,81,89,72]
}
df = pd.DataFrame(data)
print(df)
```
